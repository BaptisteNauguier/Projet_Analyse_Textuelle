```python

def mots_frequents_par_mois(dictionnaire, mois):
    # Vérifier si le mois est présent dans le dictionnaire
    if mois not in dictionnaire:
        print(f"Le mois {mois} n'est pas présent dans le dictionnaire.")
        return []

    # Extraire le texte associé au mois
    texte = dictionnaire[mois]
    
    
    # Convertir le texte en liste de mots
    mots = []
    for i in range (0,len(texte),1) :
        mots.append(texte[i].split(" "))
        
    mots = [elt for sous_liste in mots for elt in sous_liste]
    
    
    # Compter la fréquence de chaque mot
    freq = {}
    for elt in mots:
        if elt in freq:
            freq[elt] += 1
        else :
            freq[elt] = 1

    # Trier les mots par fréquence décroissante
    mots_tries = dico_trie = {k: v for k, v in sorted(freq.items(), key=lambda x: x[1], reverse=True)}

    # Retourner les 10 premiers mots les plus fréquents
    return mots_tries


mots_frequents_par_mois(dico,'Bouches-du-Rhône')
```
